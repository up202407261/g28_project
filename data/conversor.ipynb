```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('databaseFINAL.db', timeout=5)
cur = con.cursor()
df = pd.read_csv('g28_original.csv')
df
```

```python
df1 = df.loc[:, 'sponsor_id':'contribution']
df1
```

```python
for index, row in df1.iterrows():
    values = ", ".join([f'"{row_item}"' for row_item in row])
    insert_sql = f"INSERT INTO Sponsor ({', '.join(df1.columns.str.replace(' ', '_'))}) VALUES ({values})"
    cur.execute("SELECT COUNT(*) FROM Sponsor WHERE sponsor_id = ?", (row['sponsor_id'],))
    if cur.fetchone()[0] == 0:
        cur.execute(insert_sql)
```

```python
df2 = df.loc[:, 'team_name':'ranking']
df3 = df.loc[:, 'team_id']
df4 = pd.concat([df3, df2], axis=1)
df4
```

```python
for index, row in df4.iterrows():
    values = ", ".join([f'"{row_item}"' for row_item in row])
    insert_sql = f"INSERT INTO Team ({', '.join(df4.columns.str.replace(' ', '_'))}) VALUES ({values})"
    cur.execute("SELECT COUNT(*) FROM Team WHERE team_id = ?", (row['team_id'],))
    if cur.fetchone()[0] == 0:
        cur.execute(insert_sql)
```

```python
df5 = df.loc[:, 'tournament_id']
df6 = df.loc[:, 'tournament_name']
df7 = df.loc[:, 'location':'prize_pool']
df15 = df.loc[:, 'sponsor_id']
df8 = pd.concat([df5, df6, df7, df15], axis=1)
df8
```

```python
for index, row in df8.iterrows():
    values = ", ".join([f'"{row_item}"' for row_item in row])
    insert_sql = f"INSERT INTO Tournament ({', '.join(df8.columns.str.replace(' ', '_'))}) VALUES ({values})"
    cur.execute("SELECT COUNT(*) FROM Tournament WHERE tournament_id = ?", (row['tournament_id'],))
    if cur.fetchone()[0] == 0:
        cur.execute(insert_sql)
```

```python
df9 = df.loc[:, 'team_id':'result']
pks = []
id = 0
for row in df.iterrows():
    pk = id
    pks.append(pk)
    id += 1
df10 = pd.DataFrame({'game_id': pks})
df11 = pd.concat([df10, df9], axis=1)
df11
```

```python
for index, row in df11.iterrows():
    values = ", ".join([f'"{row_item}"' for row_item in row])
    insert_sql = f"INSERT INTO Game ({', '.join(df11.columns.str.replace(' ', '_'))}) VALUES ({values})"
    cur.execute("SELECT COUNT(*) FROM Game WHERE game_id = ?", (row['game_id'],))
    if cur.fetchone()[0] == 0:
        cur.execute(insert_sql)
```
